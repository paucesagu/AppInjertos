## Preprocesamiento

```python
datos = pd.read_csv('data/injertosTotales_401.csv', delimiter=';')
datos.head(100)
```

```python
datos.info()
```

```python
#modifica los valores ausentes con NaN y convierte las columnas a numéricas
datos = datos.apply(pd.to_numeric, errors='coerce')
# NOTE: Make sure that the class is labeled 'class' in the data file
datos.rename(columns={'Validez': 'class'}, inplace=True)

datos.head(50)
```

**Cambiar los valores 1 y 2 a 0 y 1 para que no haya orden donde 1 será true y 0 false**

```python
datos['Sexo'] = datos['Sexo'].map({2: 0, 1: 1})
datos['HTA'] = datos['HTA'].map({2: 0, 1: 1})
datos['DM'] = datos['DM'].map({2: 0, 1: 1})
datos['DLP'] = datos['DLP'].map({2: 0, 1: 1})
datos['APM'] = datos['APM'].map({2: 0, 1: 1})
datos['APQ'] = datos['APQ'].map({2: 0, 1: 1})
datos['AcVHC'] = datos['AcVHC'].map({2: 0, 1: 1})
datos['AcVHBc'] = datos['AcVHBc'].map({2.0: 0, 1.0: 1})
datos['Aminas'] = datos['Aminas'].map({2.0: 0, 1.0: 1})
#datos[['Sexo', 'HTA', 'DM', 'DLP', 'APM', 'APQ', 'AcVHC','AcVHBc', 'Aminas']] = datos[['Sexo', 'HTA', 'DM', 'DLP', 'APM', 'APQ', 'AcVHC','AcVHBc', 'Aminas']].map({2: 0})
datos.head(50)
```

```python
print(datos.info())
print(datos.describe())
```

```python
#valores ausentes: media o moda según la clase
for column in datos.columns:
        if column in ['IMC','GOT','GGT','BbT']:
            datos[column].fillna(datos.groupby("class")[column].transform("mean"), inplace=True)            
        if column in ['AcVHC','AcVHBc']:
            #la moda para ambas columnas es la misma independientemente de la etiqueta de clase
            datos[column].fillna(value=datos[column].mode()[0], inplace = True) 
```

```python
def evalua_dataset(X,y,clasificador,matriz=False,cv=10):
    y_pred = cross_val_predict(clasificador, X, y, cv=cv)
    
    acc =  metrics.accuracy_score(y, y_pred)
    precision = metrics.precision_score(y, y_pred)
    recall = metrics.recall_score(y, y_pred)
    f1 = metrics.f1_score(y, y_pred)
    print('accuracy: ',acc)
    print('precision: ',precision)
    print('recall: ', recall)
    print('f1: ',f1)
    
   # class_rep = classification_report(y, y_pred)
    
    if matriz:
        # Visualización de la matriz de confusión
        #print(class_rep)
        labels = y.unique()
        cm = metrics.confusion_matrix(y, y_pred, labels)
        sns.heatmap(cm, annot=True, cmap='Greens', fmt='d', xticklabels=labels, yticklabels=labels)
        plt.xlabel('Predictions')
        plt.ylabel('Actual values')
        plt.show()   
    
    return (acc,precision,recall,f1)
```

```python
def analisis_roc(X,y,clasificador,curva=False,cv=10):
    y_proba = cross_val_predict(clasificador, X, y, cv=cv, method='predict_proba')
    y_proba_clase1 =  pd.Series(y_proba[:,1])
    fprs, tprs, umbrales = metrics.roc_curve(y, y_proba_clase1)
    #print('FPRs: ', fprs[:5])
    #print('TPRs:    ', tprs[:5])
    #print('Umbrales:   ', umbrales[:5])
    auc = metrics.auc(fprs,tprs)
    print('auc: ', auc)

    if curva:
        plt.plot(fprs, tprs)
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.0])
        plt.title('Curva ROC')
        plt.xlabel('False Positive Rate (1 - especificidad)')
        plt.ylabel('True Positive Rate (sensibilidad)')
        plt.grid(True)
        plt.show()
    return auc
```

```python
def evalua_train_test(X_train,X_test,y_train,y_test,clasificador, matriz = False):
    
    clasificador.fit(X_train,y_train)
    y_test_pred = clasificador.predict(X_test)
    accuracy = metrics.accuracy_score(y_test, y_test_pred)
    precision = metrics.precision_score(y_test, y_test_pred)
    recall = metrics.recall_score(y_test, y_test_pred)
    f1 = metrics.f1_score(y_test, y_test_pred)
    y_test_proba = clasificador.predict_proba(X_test)
    y_proba_clase1 =  pd.Series(y_test_proba[:,1])
    fprs, tprs, umbrales = metrics.roc_curve(y_test, y_proba_clase1)
    auc = metrics.auc(fprs,tprs)
    #class_rep = classification_report(y_test, y_test_pred)

    if matriz:
        # Visualización de la matriz de confusión
       
        labels = y_test.unique()
        cm = metrics.confusion_matrix(y_test, y_test_pred, labels)
        sns.heatmap(cm, annot=True, cmap='Greens', fmt='d', xticklabels=labels, yticklabels=labels)
        plt.xlabel('Predictions')
        plt.ylabel('Actual values')
        plt.show() 
    return (accuracy,precision, recall, f1, auc)
```

```python
def analisis_roc_train_test(X_train,X_test,y_train,y_test,clasificador, curva = False):
    clasificador.fit(X_train,y_train)
    y_test_proba = clasificador.predict_proba(X_test)
    y_proba_clase1 =  pd.Series(y_test_proba[:,1])
    fprs, tprs, umbrales = metrics.roc_curve(y_test, y_proba_clase1)
    auc = metrics.auc(fprs,tprs)
    if curva:
        plt.plot(fprs, tprs)
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.0])
        plt.title('Curva ROC')
        plt.xlabel('False Positive Rate (1 - especificidad)')
        plt.ylabel('True Positive Rate (sensibilidad)')
        plt.grid(True)
        plt.show()
```

# Iteraciones con shuffle

```python
def iteraciones_shuffle(pipeline, cv=LeaveOneOut()):
    suma_acc=0
    suma_auc=0
    suma_precision = 0
    suma_recall = 0
    suma_f1 = 0
    n=100
    pos_mejor_shuffle = 0
    lista_shuffles = []
    for i in range(n):
        datos_shuffle=datos_one_hot.iloc[np.random.permutation(len(datos_one_hot))]
        y_shuffle = datos_shuffle['class']
        X_shuffle = datos_shuffle.drop(['class'],axis=1)

        print("Iteracion ", i, ": ")

        scores = evalua_dataset(X_shuffle,y_shuffle,pipeline,cv=cv)
        acc = scores[0]
        auc = analisis_roc(X_shuffle,y_shuffle,pipeline,cv=cv)

        lista_shuffles.append((datos_shuffle,scores,auc))

        if i==0 or (acc>= lista_shuffles[pos_mejor_shuffle][1][0] and auc>=lista_shuffles[pos_mejor_shuffle][2]):
            pos_mejor_shuffle = i

        suma_acc += scores[0]
        suma_auc += auc    
        suma_precision += scores[1]
        suma_recall += scores[2]
        suma_f1 += scores[3]   


    media_acc = suma_acc/n
    media_auc = suma_auc/n
    media_prec = suma_precision/n
    media_rec = suma_recall/n
    media_f1 = suma_f1/n
    print("Media accuracy para ", cv, ": ", media_acc)
    print("Media area under curve para ", cv, ": ", media_auc)
    print("Media precision para ", cv, ": ", media_prec)
    print("Media recall para ", cv, ": ", media_rec)
    print("Media F1 para ", cv, ": ", media_f1)


    print("Mejor shuffle: ")
    print("ACC: ", lista_shuffles[pos_mejor_shuffle][1][0])
    print("AUC: ", lista_shuffles[pos_mejor_shuffle][2])
    print("PRECISION: ", lista_shuffles[pos_mejor_shuffle][1][1])
    print("RECALL: ", lista_shuffles[pos_mejor_shuffle][1][2])
    print("F1-SCORE: ", lista_shuffles[pos_mejor_shuffle][1][3])
```

```python
def iteraciones_train_test(pipeline):
    suma_acc=0
    suma_auc=0
    suma_precision = 0
    suma_recall = 0
    suma_f1 = 0
    n=100

    for i in range(n):
        datos_shuffle=datos_one_hot.iloc[np.random.permutation(len(datos_one_hot))]
        y_shuffle = datos_shuffle['class']
        X_shuffle = datos_shuffle.drop(['class'],axis=1)
        X_train, X_test, y_train, y_test = train_test_split(X_shuffle, y_shuffle, test_size=0.20)

        print("Iteracion ", i, ": ")

        pipeline.fit(X_train,y_train)
        y_test_pred = pipeline.predict(X_test)
        acc = metrics.accuracy_score(y_test, y_test_pred)

        y_test_proba = pipeline.predict_proba(X_test)
        y_proba_clase1 =  pd.Series(y_test_proba[:,1])
        fprs, tprs, umbrales = metrics.roc_curve(y_test, y_proba_clase1)
        auc = metrics.auc(fprs,tprs)

        suma_acc += acc
        suma_auc += auc  
        suma_precision += metrics.precision_score(y_test, y_test_pred)
        suma_recall += metrics.recall_score(y_test, y_test_pred)
        suma_f1 += metrics.f1_score(y_test, y_test_pred)


    media_acc = suma_acc/n
    media_auc = suma_auc/n
    media_prec = suma_precision/n
    media_rec = suma_recall/n
    media_f1 = suma_f1/n
    print("Media accuracy para Train/Test: ", media_acc)
    print("Media area under curve para Train/Test: ", media_auc)
    print("Media precision para Train/Test: ", media_prec)
    print("Media recall para Train/Test: ", media_rec)
    print("Media F1 para Train/Test: ", media_f1)
```
